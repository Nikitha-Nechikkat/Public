```python
#Importing the csv file to the environment 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r'C:\Users\Dhanesh\OneDrive\Desktop\DSA\cars_data.csv')
print(df.info())
```

```python
df.head()
```

```python
#Check for the null values present in the dataset

print(df.isnull().sum())
```

```python
# Male vs Female buyers participating in the sale 
ax=sns.countplot(x="Buyer Gender", data=df, palette="Set3")
ax.set_title('Male Vs Female Buyers');

for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='black', size=10)
plt.show()
```
