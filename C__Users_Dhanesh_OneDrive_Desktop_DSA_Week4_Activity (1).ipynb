```python
# Reading the dataset to python environment
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r'C:\Users\Dhanesh\OneDrive\Desktop\DSA\Telco-Customer-Churn.xlsx')
print(df.info())
```

```python
#Check for the null values present in the dataset
print(df.isnull().sum())
```

```python
df.head()
```

```python
#Q1 Compare churn count with respect to gender. 
ax=sns.countplot(x="Churn", data=df,hue="gender", palette="Set3")
ax.set_title('Churn count');

for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='black', size=10)
plt.show()

#Observation 1 : 939 female and 930 male customers out of the total sample said yes to churn.
```

```python
#Q2 Find out how many female senior citizens there in the dataset
ax=sns.countplot(x="SeniorCitizen", data=df, palette="Set3",hue="gender")
ax.set_title('Male vs Female');

for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='black', size=10)
plt.show()

#Observation 2: There is 568 female Senior citizen customers in the sample taken.
```

```python
#Q3 Compare 'tenure' with 'Total Charges' 
df=pd.read_excel(r'C:\Users\Dhanesh\OneDrive\Desktop\DSA\Telco-Customer-Churn.xlsx',na_values=[" "])
plt.figure(figsize=(6,4))
sns.lineplot(x='tenure',y='TotalCharges',data = df)
plt.title("Tenure v/s Total Charges",fontsize=10)
plt.show()

# Tenure directly proportional to Total Charges
```

```python
#Q4 Find out which contract preffered by the senior citizen.
ax=sns.countplot(x="Contract", data=df, palette="Set3",hue="SeniorCitizen")
ax.set_title('Contract preffered by the citizens');
plt.xticks(rotation=45)

for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='black', size=10)
plt.show()

#Observation 4: Majority of the Senior Citizen customers prefer 'Month to Month' payment contract.
```

```python
#Q5 Comment your finds on Payment Method?
ax=sns.countplot(x="PaymentMethod", data=df, palette="Set3")
ax.set_title('Payement Method');
plt.xticks(rotation=45)

for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='black', size=10)
plt.show()

#Observation 5:
# 33% of the total sample preferred 'Ã‰lectronic checks' for payment.
# 22.8 % preferred 'Mailed check' for payment.
# 21.9 % preferred 'Bank Transfer' for the payment and the rest used 'Ã‡redit cards' for payment
```
